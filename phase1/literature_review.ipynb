## 1. Introduction to the Topic <a id='1'></a>

The global expansion of educational systems — driven by digitisation, distance learning mandates, and the pressure to serve ever-larger student cohorts — has created a structural mismatch between assessment demand and the available grading bandwidth of human educators. Manual grading of handwritten student responses is not only time-intensive but is demonstrably susceptible to **inter-rater inconsistency**, **cognitive fatigue**, and **implicit evaluator bias** (Ramsey & Duffy, 2016). Automated grading systems promise to resolve this mismatch by applying computational intelligence to the assessment pipeline at scale, with the added benefit of producing consistent, auditable, and near-instantaneous feedback.

Yet, unlike automated systems for typed or digital text — where input is already in a machine-readable form — **handwritten answer evaluation** imposes a far more demanding set of sub-problems. The system must:

1. **Perceive and interpret ink on paper** (scanned image → text)
2. **Extract linguistic meaning** from the recognised text
3. **Evaluate semantic content** against a reference answer or rubric

Each stage carries its own failure modes: degraded image quality, writer-specific script idiosyncrasies, domain-specific vocabulary, and the inherent subjectivity of natural language evaluation.

### Why It Matters

| Problem | Impact |
|---|---|
| Manual grading is slow | Delays feedback; reduces learning loop effectiveness |
| Grader inconsistency | Unfair outcomes; high inter-rater variance |
| Handwritten OCR is hard | Cursive variability, noise, layout disorder |
| No semantic understanding | Keyword graders penalise correct paraphrases |
| No end-to-end pipeline | Research treats OCR and grading as separate problems |

This literature review surveys the state of the art across each sub-problem and synthesises their intersection as it pertains to the design of an **end-to-end Automated EdTech Grading Assistant**.

## 2. Background Concepts <a id='2'></a>

### 2.1 Optical Character Recognition (OCR)

**Optical Character Recognition (OCR)** refers to the computational process of converting raster images of text — whether printed, typewritten, or handwritten — into machine-encoded symbols.

**Classical OCR Pipeline:**
```
Image → Binarisation → Segmentation → Feature Extraction → Classification → Text
```

Classical systems used:
- **Hu moments**, **HOG descriptors**, **Zernike moments** for feature extraction
- **Template matching** or **k-NN** for character classification (Mori et al., 1999)

> ⚠️ These approaches achieved high accuracy on printed, regular-font documents but were fundamentally brittle when confronted with the variability of handwritten script.

Modern OCR has migrated to **deep learning architectures** that jointly learn features and classification boundaries from large annotated corpora.

### 2.2 Handwritten Text Recognition (HTR)

**HTR** is the sub-discipline of OCR focused on the unique challenges of human cursive or semi-cursive script.

**Principal Difficulties:**

| Challenge | Description |
|---|---|
| Inter-writer variability | Different people form the same letter very differently |
| Intra-writer variability | Same writer varies across fatigue, emotion, instrument |
| Ligature ambiguity | Character boundaries unclear or absent in cursive |
| Context dependency | Same stroke = different character depending on neighbours |

**Standard Benchmarks:**
- **IAM Handwriting Database** (Marti & Bunke, 2002) — English sentences
- **RIMES corpus** (Augustin et al., 2006) — French handwritten mail
- **Washington Historical Manuscript** (Fischer et al., 2012)

**Evaluation Metrics:**
- **CER** (Character Error Rate) = edit distance / reference length
- **WER** (Word Error Rate) = word-level edit distance / reference length

### 2.3 NLP & Semantic Similarity

For automated grading, the key NLP task is **Semantic Textual Similarity (STS)**: given a student answer and a reference answer, compute a score reflecting the degree to which they convey the same meaning.

**Evolution of STS approaches:**

```
BLEU / ROUGE (lexical overlap)
        ↓
TF-IDF cosine similarity
        ↓
Word2Vec / GloVe sentence averages
        ↓
BERT contextualised embeddings (Devlin et al., 2018)
        ↓
Sentence-BERT / SBERT (Reimers & Gurevych, 2019)  ← Current SOTA
```

> 🔑 **Key insight:** Early lexical methods penalise semantically equivalent paraphrases. Transformer-based models capture meaning as a function of context — critical for grading student answers that use different wording than the model answer.

### 2.4 Document Layout Analysis (DLA)

Real-world student answer sheets are **not** clean, single-column text documents. They contain:
- Question numbers and answer boxes
- Margin annotations and diagrams
- Multi-column arrangements
- Struck-through text and corrections

**Document Layout Analysis (DLA)** decomposes a page image into structural regions (headers, body text, figures, tables) *before* OCR/HTR processing, ensuring recognised text is attributed to the correct question.

**DLA Approaches:**
- Rule-based zone detection (heuristic, fast)
- Mask R-CNN applied to page segmentation (He et al., 2017)
- Graph Neural Networks for relational layout (Shen et al., 2021)

### 2.5 AI-Based Automated Evaluation Systems

Automated grading systems span a spectrum:

| Era | Approach | Example |
|---|---|---|
| 1960s | Surface features (word count, sentence length) | PEG — Page (1966) |
| 2000s | LSA + syntactic matching | Mohler et al. (2011) |
| 2016–2019 | CNN/RNN + attention for essay scoring | Dong & Zhang (2016) |
| 2020+ | BERT/Transformer end-to-end | Uto et al. (2021); Wang et al. (2023) |

> The transition from **handcrafted feature engineering** to **representation learning** is the primary axis of progress in this literature.

## 3. Review of Existing Work <a id='3'></a>

### 3.1 Foundational HTR Architectures

---

#### 📄 Paper 1 — Graves et al. (2006) — CTC + LSTM

**Full Title:** *Connectionist Temporal Classification: Labelling Unsegmented Sequence Data with Recurrent Neural Networks*  
**Venue:** ICML 2006

**Approach:**  
Introduced **Connectionist Temporal Classification (CTC)** as a training objective for bidirectional LSTMs applied to HTR. The model processes a sequence of image column features and emits character probabilities without requiring explicit character-level boundary annotations — eliminating the need for manual word segmentation that had bottlenecked earlier HMM-based systems.

```
Image columns → BiLSTM encoder → CTC decoder → Character sequence
```

**Dataset:** IAM Handwriting Database  
**Key Result:** WER = 25.9% on IAM test set (significant improvement over GMM-HMM baselines at the time)

**Limitations:**
- Requires large quantities of character-level labelled data
- Slow inference due to sequential LSTM computation
- Does not handle variable-length full document pages natively

---

#### 📄 Paper 2 — Shi, Bai & Yao (2016) — CRNN

**Full Title:** *An End-to-End Trainable Neural Network for Image-Based Sequence Recognition*  
**Venue:** IEEE TPAMI

**Approach:**  
Proposed the **Convolutional Recurrent Neural Network (CRNN)** — a hybrid architecture coupling CNN feature extraction (VGG-derived) with BiLSTM sequence modelling and CTC decoding.

```
Word image → CNN (feature maps) → Column sequence → BiLSTM → CTC → Text
```

**Dataset:** IIIT-5K-Word, SVT, ICDAR  
**Key Result:** Accuracy = 97.8% on IIIT-5K with lexicon

**Limitations:**
- Performance degrades significantly on unconstrained handwriting
- Assumes relatively uniform, isolated word images
- Does not generalise to mixed printed-handwritten documents

---

#### 📄 Paper 3 — Bluche & Messina (2017) — Gated CNN for HTR

**Full Title:** *Gated Convolutional Recurrent Neural Networks for Multilingual Handwriting Recognition*  
**Venue:** ICDAR 2017

**Approach:**  
Extended CRNN with **gated convolutional layers** and **multi-dimensional LSTM units** to process full text lines rather than isolated word patches, leveraging broader horizontal context for cursive ligature resolution.

**Dataset:** IAM, RIMES  
**Key Result:** CER = 3.2% on RIMES (line level)

**Limitations:**
- Still line-level; does not address document-level layout
- Cannot parse spatially disordered answer sheets
- High training data requirements

### 3.2 Transformer-Based Handwriting Recognition

---

#### 📄 Paper 4 — Li et al. (2021) — TrOCR

**Full Title:** *TrOCR: Transformer-Based Optical Character Recognition with Pre-Trained Models*  
**Venue:** arXiv:2109.10282

**Approach:**  
Treats OCR as a **sequence-to-sequence image-captioning** task. A Vision Transformer (ViT/BEiT) encodes the word image into patch embeddings; a RoBERTa-based decoder autoregressively generates the character sequence. Pre-training on large synthetic and real-world datasets enables state-of-the-art results without CTC.

```
Word image → ViT patch embeddings → Transformer decoder (RoBERTa) → Text
```

**Dataset:** IAM (handwritten), SROIE (printed receipts)  
**Key Result:** CER = 2.89% on IAM word-level; CER = 2.60% on IAM sentence-level (large model)

**Limitations:**
- Substantial computational requirements (ViT-Large + RoBERTa-Large)
- Autoregressive decoding is slow for high-throughput inference
- Performance degrades on spatially disordered layouts beyond isolated lines

---

#### 📄 Paper 5 — Kang et al. (2022) — Position-Aware HTR Transformer

**Full Title:** *Pay Attention to What You Read: Non-Recurrent Handwritten Text-Line Recognition*  
**Venue:** Pattern Recognition, 2022

**Approach:**  
Combines patch-based image tokenisation with a text decoder conditioned on **both spatial position and visual features**, enabling handling of variable-length text lines with complex layout structures. Strong generalisation to degraded manuscripts — paralleling noisy scanned exam sheets.

**Dataset:** IAM, Saint Gall, Parzival historical manuscripts  
**Key Result:** WER = 16.8% on Saint Gall

**Limitations:**
- Specialised for historical scripts; requires domain-specific fine-tuning for modern student handwriting
- Computational overhead comparable to TrOCR

### 3.3 Text Detection & Document Layout Analysis

---

#### 📄 Paper 6 — Baek et al. (2019) — CRAFT

**Full Title:** *Character Region Awareness for Text Detection*  
**Venue:** CVPR 2019

**Approach:**  
CRAFT regresses per-pixel **heatmaps** representing the likelihood of individual character centres and the affinity between adjacent characters. Enables detection of arbitrarily shaped and oriented text regions — a significant advantage for handwriting where text lines follow non-linear trajectories.

**Dataset:** ICDAR 2015, Total-Text, MSRA-TD500  
**Key Result:** F1 = 86.3% on Total-Text (curved text)

**Limitations:**
- Detection-only; no recognition module
- Performance on handwritten exam sheets with overlapping strokes not formally evaluated

---

#### 📄 Paper 7 — Shen et al. (2021) — GNN Document Layout Analysis

**Full Title:** *LayoutParser: A Unified Toolkit for Deep Learning Based Document Image Analysis*  
**Venue:** ICDAR 2021

**Approach:**  
Frames document layout analysis as a **graph neural network** problem — page elements as graph nodes, structural relationships as edges. The relational approach outperforms region-based convolutional detectors on complex multi-column layouts and naturally encodes reading order.

**Dataset:** PubLayNet, DocBank  
**Key Result:** mAP = 93.8% on PubLayNet

**Limitations:**
- Evaluated only on typeset academic documents
- Applicability to handwritten examination papers not demonstrated

---

#### 📄 Paper 8 — Ye & Doermann (2015) — Text Detection Survey

**Full Title:** *Text Detection and Recognition in Imagery: A Survey*  
**Venue:** IEEE TPAMI 2015

**Approach:**  
Comprehensive survey cataloguing the progression from classical morphological operations through deep learning. Identifies three key remaining challenges:
1. Scene complexity and background clutter
2. Perspective distortion in naturally photographed documents
3. Segmentation of touching/overlapping characters in cursive handwriting

> These challenges remain only partially resolved and are acutely relevant to the grading of student-produced answer sheets.

### 3.4 Semantic Similarity Models for Text Grading

---

#### 📄 Paper 9 — Reimers & Gurevych (2019) — Sentence-BERT (SBERT)

**Full Title:** *Sentence-BERT: Sentence Embeddings using Siamese BERT-Networks*  
**Venue:** EMNLP 2019

**Approach:**  
Modifies BERT using **siamese and triplet network structures** to derive semantically meaningful fixed-size sentence embeddings efficiently. Semantic similarity is computed as cosine distance between pre-computed vectors — enabling real-time grading without quadratic pairwise BERT computation.

```
Student answer  → BERT encoder → Pooled embedding ┐
                                                    ├─ cosine similarity → score
Reference answer → BERT encoder → Pooled embedding ┘
```

**Dataset:** STS-Benchmark (evaluation), NLI datasets (training)  
**Key Result:** Pearson r = 0.896 on STS-B; ~65× faster than raw BERT for all-pairs search

**Limitations:**
- Pre-trained on general-domain text; domain shift on educational vocabulary (chemistry, economics) may degrade performance without fine-tuning
- Short, telegraphic student answers carry high embedding variance

---

#### 📄 Paper 10 — Mohler et al. (2011) — Short Answer Grading with LSA

**Full Title:** *Learning to Grade Short Answer Questions using Semantic Similarity Measures and Dependency Graph Alignments*  
**Venue:** ACL 2011

**Approach:**  
Combined **Latent Semantic Analysis (LSA)** with **syntactic dependency graph matching** for short-answer grading. The combination of distributional semantics and structural matching outperformed pure lexical or LSA-only baselines.

**Dataset:** Mohler & Mihalcea (2009) short-answer corpus; SemEval-2013 Task 7  
**Key Result:** Pearson r = 0.518–0.590 with human scores

**Limitations:**
- LSA is count-based and non-contextual
- Assumes typed text input — no OCR component whatsoever
- Graph-based matching is computationally expensive at scale

---

#### 📄 Paper 11 — Uto et al. (2021) — BERT-IRT Hybrid AES

**Full Title:** *Neural Automated Essay Scoring Incorporating Handcrafted Features*  
**Venue:** COLING 2021

**Approach:**  
Combined **BERT-based essay encoding** with **Item Response Theory (IRT)** to jointly model essay quality and rater characteristics. Explicitly represents rater severity and consistency as latent variables — directly reducing the impact of grader bias.

**Dataset:** ASAP Essay Scoring Corpus (Kaggle)  
**Key Result:** QWK = 0.81 on ASAP prompt 2

**Limitations:**
- Typed essay text only — no OCR integration
- IRT calibration requires a large corpus of multiply-rated responses

---

#### 📄 Paper 12 — Dong & Zhang (2016) — Hierarchical RNN for AES

**Full Title:** *Automatic Features for Essay Scoring: An Empirical Study*  
**Venue:** EMNLP 2016

**Approach:**  
Hierarchical RNN with **attention mechanisms** at both word and sentence levels for automatic essay scoring. The attention mechanism identifies and weights the most grade-relevant segments. Surpassed feature-engineered AES baselines without any manual feature design.

**Dataset:** ASAP corpus  
**Key Result:** Accuracy = 89.4%

**Limitations:**
- Cannot handle multi-paragraph text without memory constraints
- Not designed for short-form or keyword-anchored answers typical of handwritten worksheets

### 3.5 Integrated Automated Grading Systems

---

#### 📄 Paper 13 — Tashu & Horváth (2020) — End-to-End Handwritten Answer Grading

**Full Title:** *Off-Line Handwritten Text Recognition with Multi-Task Learning*  
**Venue:** ICAART 2020

**Approach:**  
One of the most directly relevant systems: an end-to-end pipeline for grading handwritten exam answers using a **CNN-LSTM** architecture for recognition followed by a **semantic matching** stage. Attention mechanisms handle variable-length answers; the grading stage compares recognised text to model answers using a trained similarity model.

**Dataset:** Custom university exam corpus *(not publicly available)*  
**Key Result:** Overall grading accuracy = 82%

**Limitations:**
- Non-public dataset — limits reproducibility and external validity
- Not tested on noisy, low-quality scans or photographed sheets
- Semantic matching stage uses simple similarity metrics without contextual representations

---

#### 📄 Paper 14 — Wang et al. (2023) — LLMs for Educational Assessment

**Approach:**  
Investigated GPT-4 and LLaMA-2 as **zero-shot and few-shot graders** for short-answer educational assessments. LLMs can achieve near-human grading consistency with well-designed rubric prompts — but only for digital typed text.

**Dataset:** SciEntsBank, Beetle short-answer corpora  
**Key Result:** QWK = 0.77–0.84 in few-shot settings

**Limitations:**
- No handwriting or document processing — requires typed text input
- LLM inference cost is prohibitive for large-scale batch assessment
- Zero-shot performance sensitive to prompt design and domain vocabulary

---

#### 📄 System 15 — EasyOCR & PaddleOCR (2020) — Open-Source OCR Engines

**EasyOCR** (Jaided AI): CRAFT-based text detector + CRNN/TrOCR recogniser; 80+ languages  
**PaddleOCR** (Baidu): DB detector + SVTR recogniser + built-in document layout analysis

**Informal evaluation on educational handwriting:**
- PaddleOCR performs better on structured forms
- EasyOCR is more robust on isolated word-level recognition

**Limitations:**
- Neither system fine-tuned for educational handwriting
- Both struggle with highly cursive, idiosyncratic student scripts
- No semantic grading capabilities included

### Key Comparative Findings

**1. Traditional OCR vs Transformer-Based OCR:**
- CRNN-based systems (Shi et al., 2016) offer faster inference with lower hardware requirements but show significant performance degradation on variable student handwriting.
- TrOCR (Li et al., 2021) achieves superior recognition accuracy but requires substantially greater computational resources.
- **Opportunity:** A hybrid approach — TrOCR for high-ambiguity segments, lightweight CRNN for confident regions — represents a promising middle ground **not explored** in existing literature.

**2. Rule-Based vs Semantic Grading:**
- Keyword matching is fast and interpretable but fails on paraphrased correct answers and synonymous terminology.
- SBERT-based cosine similarity is more semantically robust but requires threshold calibration.
- The IRT-augmented approach of Uto et al. (2021) is promising for bias correction but has **not been combined** with an OCR input stage.

**3. The Integration Gap:**
> No reviewed system coherently integrates document preprocessing, handwriting recognition, and semantic grading into a single, robustly evaluated pipeline.

This literature review has traced the development of automated handwritten grading systems from their classical OCR and rule-based foundations to the current frontiers of transformer-based recognition and neural semantic evaluation.

### Why This Project is Necessary

> **Critical gaps persist in five dimensions:**
> 1. Absence of end-to-end integrated pipelines
> 2. Insufficient robustness to real-world scan degradation  
> 3. Semantic limitations of short-answer grading models  
> 4. Layout-unawareness of existing HTR systems  
> 5. Complete absence of student-facing diagnostic feedback in any reviewed system  

The **Automated EdTech Grading Assistant** proposed in this project addresses each of these gaps through a principled, modular pipeline architecture combining state-of-the-art HTR, layout analysis, and semantic similarity modelling with novel confidence-gated routing and topic-level feedback aggregation.
